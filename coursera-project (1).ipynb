```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mlt
from sklearn import preprocessing
```

```python
df=pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/loan_train.csv')
```

```python
df
```

```python
data=df.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1)
data['customer']=range(0,len(df))


data=data[['customer','age','Gender','terms','Principal','effective_date','due_date','education','loan_status']]
data
```

```python
from sklearn.preprocessing import LabelEncoder
num=LabelEncoder()
data['Gender']=num.fit_transform(data['Gender'].astype(str))
data['loan_status']=num.fit_transform(data['loan_status'].astype(str))
data['education']=num.fit_transform(data['education'].astype(str))
data
```

```python
X=data[['age','Gender','terms','Principal','education']].values
y=data['loan_status'].values
X[0:5]
#y[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=3)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
```

```python
from sklearn.metrics import jaccard_score
from sklearn.metrics import log_loss
from sklearn.metrics import f1_score
from sklearn import metrics

def JS(y_test,yhat,k):
    js=jaccard_score(y_test, yhat,pos_label=k)
    return js
def F1(y_test,yhat):
    f1=f1_score(y_test, yhat, average='weighted') 
    return f1
def LoL(y_test,yhat_prob):
    lol=log_loss(y_test, yhat_prob)
    return lol
```

```python
# K NEIGHBORS CLASSIFIER

Ks = 10
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])
    
k = mean_acc.argmax()+1 #k for max accuracy
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)

yhat_knc=neigh.predict(X_test)
yhat_prob_knc=neigh.predict_proba(X_test)

```

```python
# ACCURACY FOR K NEIGHBORS CLASSIFIER

print('Jcard Score of K NEIGHBORS classifier is',JS(y_test,yhat_knc,1))
print('F1 of K NEIGHBORS classifier is',F1(y_test,yhat_knc))
print('Logloss of K NEIGHBORS classifier is',LoL(y_test,yhat_prob_knc))
```

```python
# DECISION TREE CLASSIFIER
LoanTree = DecisionTreeClassifier(criterion="gini", max_depth = 4)
LoanTree.fit(X_train,y_train)
preTree = LoanTree.predict(X_test)
preTree_prob = LoanTree.predict_proba(X_test)
```

```python
# ACCURACY OF DECISION TREE

print('Jcard Score of DECISION TREE classifier is',JS(y_test,preTree,1))
print('F1 of DECISION TREE classifier is',F1(y_test,preTree))
print('Logloss of DECISION TREE is',LoL(y_test,preTree_prob))
```

```python
# LOGISTIC REGRESSION

LR=LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
yhat_lr=LR.predict(X_test)
yhat_prob_lr=LR.predict_proba(X_test)

```

```python
# ACCURACY OF LOGISTIC REGRESSION

print('Jcard Score of LOGISTIC REGRESSION classifier is',JS(y_test,yhat_lr,1))
print('F1 of LOGISTIC REGRESSION classifier is',F1(y_test,yhat_lr))
print('Logloss of LOGISTIC REGRESSION is',LoL(y_test,yhat_prob_lr))
```

```python
# SUPPORT VECTOR MACHINE

sv = svm.SVC(kernel='rbf')
sv.fit(X_train,y_train)

yhat_svm=sv.predict(X_test)
yhat_prob_svm=sv.predict(X_test)
```

```python
# ACCURACY OF SVM

print('Jcard Score of SUPPORT VECTOR MACHINE classifier is',JS(y_test,yhat_svm,1))
print('F1 of SUPPORT VECTOR MACHINE classifier is',F1(y_test,yhat_svm))
print('Logloss of STANDARD VECTOR MACHINE is',LoL(y_test,yhat_prob_svm))
```
